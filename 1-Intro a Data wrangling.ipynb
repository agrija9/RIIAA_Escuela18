### Clean Data

#### Replace label column to numerical values

```python
mapping_II = {'s': 0, 'b': 1} # signal = boson,  background = any noise
ds = ds.replace({'Label': mapping_II}) # Mapea el nombre de la especie a formato numérico
ds
```

#### Explore Higgs Boson Data

```python
sub_ds = ds.sample(n=100)
print(sub_ds.shape)
```

```python
# Creamos un subconjunto de columnas del data set
sns.pairplot(sub_ds[['DER_mass_MMC', 'DER_mass_transverse_met_lep', 'Weight', 'Label']], hue='Label')
```
